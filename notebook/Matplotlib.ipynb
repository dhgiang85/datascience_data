# **Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,20)
y = x*x
z = y+x
```

```python
plt.plot(x,z,x,y,x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('example')
plt.show()
```

```python
# Subplot
plt.subplot(1,2,1)
plt.plot(x,x,'r')

plt.subplot(1,2,2)
plt.plot(x,y)
plt.tight_layout()
```

```python
# Type of plots
plt.subplot(2,2,1)
plt.scatter(x,y)

plt.subplot(2,2,2)
plt.bar(x,y)

plt.subplot(2,2,3)
plt.plot(x,y)

plt.subplot(2,2,4)
plt.fill(x,y)
plt.tight_layout()
```

```python
plt.polar(x,y)
```

```python
plt.plot(x,z,label ='first plot')
plt.legend(loc=1)
```

## **Object oriented plots**

```python
fig = plt.figure() 
x1 = fig.add_axes([0.1,0.1,0.8,0.8])
x1.plot(x,y)
x2 = fig.add_axes([0.4,0.4,0.3,0.3])
x2.plot(y,x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(x,x)
fig.tight_layout()
```

```python
# figure size
fig = plt.figure(figsize =(5,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1,figsize =(5,5))
axes[0].plot(x,y,marker='o',markersize=5,markerfacecolor='b',markeredgecolor='pink')
axes[0].set_ylabel('y-axis')
axes[0].set_title('example')
axes[1].plot(x,z,color='r',lw=3,alpha =0.6,linestyle=':')
axes[1].set_ylabel('z-axis')
axes[1].set_xlabel('x-axis')
axes[1].set_xlim([2,12])
fig.tight_layout()
```
